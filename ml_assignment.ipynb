Question 1: What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.

Answer:
* AI (Artificial Intelligence): Broad field where machines perform tasks like humans (decision-making, problem-solving).
* ML (Machine Learning): Subset of AI where machines learn from data without being explicitly programmed.
* DL (Deep Learning): Subset of ML using neural networks with many layers for complex tasks like vision & speech.
* Data Science: Focuses on data handling, analysis, statistics, insights & visualizations.
* Scope difference: AI → Big picture, ML → learning models, DL → advanced neural learning, DS → extracting value from data.

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

Answer:
* Overfitting: Model learns noise, performs well on training but poorly on testing.
* Underfitting: Model is too simple, performs poorly on both training & testing.
* Detection: Compare train vs test error

  (large gap → overfitting, high error both → underfitting).
* Prevention: Cross-validation, regularization, pruning, early stopping.
* Bias-variance tradeoff: Find balance between overly simple and overly complex models.


Question 3: How would you handle missing values in a dataset? Explain at least three methods with examples

Answer:
* Deletion: Remove rows/columns if missing data is small (e.g., <5%).
* Mean/Median Imputation: Replace missing numeric values with average/median of column.
* Mode Imputation: For categorical data, replace with most frequent category.
* Predictive Imputation: Use ML models (e.g., KNN) to predict missing values.
* Flagging: Add a new “missing” indicator feature if missing patterns matter.


Question 4: What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).

Answer:
* When one class has very few samples vs another (e.g., fraud 1% vs non-fraud 99%).
* Problem: Model becomes biased toward majority class.
* Oversampling: Increase minority class using SMOTE, Random Oversampling.
* Undersampling: Reduce majority class to balance data.
* Class Weights: Penalize misclassification of minority class in model training.

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Answer:
* Different features have different scales → model gets biased toward bigger values.
* Min-Max Scaling: Converts values to range [0,1] → good for KNN, Neural Nets.
* Standardization (Z-score): Data becomes mean=0, std=1 → preferred for SVM, Logistic Regression.
* Helps gradient descent converge faster.
* Essential for distance-based and optimization-based algorithms.


Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

Answer:
* Label Encoding: Converts categories to numbers (0,1,2…) → use for ordinal data (e.g., small < medium < large).
* One-Hot Encoding: Creates separate binary columns → use for nominal data (no order like colors).
* Prevents model from assuming wrong order for nominal features.
* One-Hot increases dimensionality, especially with many categories.
* Label Encoding is simple but risky for unordered factors.


```python
# Question 7: Google Play Store Dataset
# a). Analyze the relationship between app categories and ratings.
# Which categories have the highest/lowest average ratings,
# and what could be the possible reasons?
# Dataset: https://github.com/MasteriNeuron/datasets.git
# (Include your Python code and output in the code box below.)
# Answer:

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('googleplaystore.csv')

# Remove missing and incorrect values
df = df[['Category', 'Rating']].dropna()
df = df[df['Rating'] <= 5]  # filter invalid ratings

# Calculate average rating per category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Average Rating by Category:\n")
print(category_ratings)

# Plot (optional for visualization)
plt.figure(figsize=(10,8))
category_ratings.plot(kind='bar')
plt.title('Average App Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# Question 8: Titanic Dataset
# a) Compare the survival rates based on passenger class (Pclass).
# Which class had the highest survival rate, and why do you think that happened?
# b) Analyze how age (Age) affected survival.
# Group passengers into children (Age < 18) and adults (Age ≥ 18).
# Did children have a better chance of survival?
# Dataset: https://github.com/MasteriNeuron/datasets.git
# (Include your Python code and output in the code box below.)
# Answer:

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("titanic.csv")   # Update file name if different

# Clean data (remove NaN age only when needed later)
df_survival = df[['Survived', 'Pclass']]

# (A) Survival Rate by Passenger Class
survival_by_class = df_survival.groupby('Pclass')['Survived'].mean()
print("\nSurvival Rate by Passenger Class:\n", survival_by_class)

# Bar plot for Pclass vs Survival
plt.figure(figsize=(6,4))
survival_by_class.plot(kind='bar')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.tight_layout()
plt.show()

# (B) Age-based Survival
df_age = df[['Survived', 'Age']].dropna()
df_age['Group'] = df_age['Age'].apply(lambda x: "Child" if x < 18 else "Adult")

survival_by_age = df_age.groupby('Group')['Survived'].mean()
print("\nSurvival Rate by Age Group:\n", survival_by_age)

# Bar plot for Age Group vs Survival
plt.figure(figsize=(6,4))
survival_by_age.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Survival Rate: Children vs Adults")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.tight_layout()
plt.show()
```

```python
# Question 9: Flight Price Prediction Dataset
# a) How do flight prices vary with the days left until departure?
# Identify any exponential price surges and recommend the best booking window.
# b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai).
# Which airlines are consistently cheaper/premium, and why?
# Dataset: https://github.com/MasteriNeuron/datasets.git
# (Include your Python code and output in the code box below.)
# Answer:

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("flight_price.csv")    # Update filename if different

# Cleaning
df = df[['airline', 'source_city', 'destination_city', 'price', 'days_left']].dropna()

# (A) Relationship: Days Left vs Price
avg_price_by_days = df.groupby('days_left')['price'].mean()

print("\nAverage Price by Days Left:\n")
print(avg_price_by_days.head(20))  # Just printing first 20 for preview

# Plot price trend
plt.figure(figsize=(8,5))
avg_price_by_days.plot()
plt.title("Flight Price vs Days Left")
plt.xlabel("Days Left Before Departure")
plt.ylabel("Average Price")
plt.tight_layout()
plt.show()

# (B) Airline Comparison for Delhi → Mumbai
route_df = df[(df['source_city']=='Delhi') & (df['destination_city']=='Mumbai')]
airline_prices = route_df.groupby('airline')['price'].mean().sort_values()

print("\nDelhi → Mumbai Route Price Comparison:\n")
print(airline_prices)

# Bar plot
plt.figure(figsize=(8,5))
airline_prices.plot(kind='bar')
plt.title("Airline Price Comparison (Delhi → Mumbai)")
plt.xlabel("Airline")
plt.ylabel("Average Ticket Price")
plt.tight_layout()
plt.show()
```

```python
# Question 10: HR Analytics Dataset
# a). What factors most strongly correlate with employee attrition?
# Use visualizations to show key drivers (e.g., satisfaction, overtime, salary).
# b). Are employees with more projects more likely to leave?
# Dataset: hr_analytics
# Answer:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hr_analytics.csv")

# Rename key columns for easier usage
df.rename(columns={'left': 'Attrition', 'number_project': 'Projects'}, inplace=True)

# Convert Attrition to 0/1 format (already 0/1 in most versions)
df['Attrition'] = df['Attrition'].astype(int)

print("\nDataset Columns:", df.columns)

# A) Correlation Analysis for Attrition
corr_factors = ['satisfaction_level', 'average_montly_hours', 'time_spend_company',
                'promotion_last_5years', 'Work_accident', 'Projects', 'Attrition']

corr_matrix = df[corr_factors].corr()['Attrition'].sort_values(ascending=False)
print("\nCorrelation with Attrition:\n", corr_matrix)

# Visualization 1: Satisfaction Level vs Attrition
plt.figure(figsize=(10,5))
sns.boxplot(x='Attrition', y='satisfaction_level', data=df)
plt.title("Satisfaction Level vs Attrition")
plt.show()

# Visualization 2: Overtime-style factor: average monthly working hours
plt.figure(figsize=(10,5))
sns.boxplot(x='Attrition', y='average_montly_hours', data=df)
plt.title("Working Hours vs Attrition")
plt.show()

# Visualization 3: Salary vs Attrition
plt.figure(figsize=(8,5))
sns.countplot(x='salary', hue='Attrition', data=df)
plt.title("Salary vs Attrition")
plt.show()

# B) Do more projects cause attrition?
plt.figure(figsize=(10,5))
sns.countplot(x='Projects', hue='Attrition', data=df)
plt.title("Projects vs Attrition")
plt.show()

projects_attrition = df.groupby('Projects')['Attrition'].mean()
print("\nAttrition Rate by Number of Projects:\n", projects_attrition)
```
